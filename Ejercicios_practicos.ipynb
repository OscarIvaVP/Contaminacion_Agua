# **Ejercicio 1. Carga Contaminante**

Hay dos empresas que están ubicadas muy cerca entre estas y tienen puntos de vertimiento continuos en el río Guayuriba (Figura). Una de estas empresas es de la industria láctea que tiene un vertimiento con caudal de 410 litros/segundo y una concentración de 21 mg/litro de DBO, mientras que la otra empresa es de la industria cárnica tiene un vertimiento con un caudal de 0.15 litros/segundo y una concentración de 60000 mg/litro de DBO ¿Qué industria es la que mayor carga contaminante  en kg tiene sobre el río al día?

image.png

**Modelo simple**

```python
Q1 = 222   # L/Seg
C1 = 12    # mg/L
Q2 = 10.5  # L/seg
C2 = 4000 # MG/L

W1 = Q1*C1*(1/1000000)*(60*60*24)
W2 = Q2*C2*(1/1000000)*(60*60*24)

if W1 > W2:
  print(f"La industria láctea tiene la mayor carga contaminante con {W1:.2f} kg/día, teniendo en cuenta que la industria cárnica tiene una carga contaminante de {W2:.2f} kg/día")
else:
  print(f"La industria cárnica tiene la mayor carga contaminante con {W2:.2f} kg/día, teniendo en cuenta que la industria láctea tiene una carga contaminante de {W1:.2f} kg/día")
```

**Modelo dinámico**

```python
Q1 = float(input("Ingrese el caudal de la industria láctea en litros/segundo:  "))
C1 = float(input("Ingrese la concentración del contaminante de la industria láctea en mg/litros:  "))
Q2 = float(input("Ingrese el caudal de la industria cárnica en litros/segundo:  "))
C2 = float(input("Ingrese la concentración del contaminante de la industria cárnica en mg/litros:  "))

W1 = Q1*C1*(1/1000000)*(60*60*24)
W2 = Q2*C2*(1/1000000)*(60*60*24)

print(" ")

if W1 > W2:
  print(f"La industria láctea tiene la mayor carga contaminante con {W1:.2f} kg/día, teniendo en cuenta que la industria cárnica tiene una carga contaminante de {W2:.2f} kg/día")
else:
  print(f"La industria cárnica tiene la mayor carga contaminante con {W2:.2f} kg/día, teniendo en cuenta que la industria láctea tiene una carga contaminante de {W1:.2f} kg/día")
```

# **Ejercicio 2.**

En la ciudad de Villavicencio, está situada una planta extractora de aceite de palma, la cual tiene un efluente residual que es vertido sobre el río  Ocoa, tal y como se observa en la figura 1. Luego de realizar el monitoreo de calidad del agua en este tramo del río, se determinó que existen una concentración de 9 mg/l sodio CS, A y un caudal QA = 25 m3/s de agua en el punto A del río. Por otro lado, se determino que la concentración del sodio CS, W en el agua residual es de 300 mg/l, y el caudal QW = 10 m3/s. Asumiendo que ocurre una mezcla completa. Determine:

a) La concentración del sodio en el punto B.
b) Indicar ¿Qué ocurrirá si en época de verano se disminuye el caudal del río en el punto A al 10%?
c) Describir que tipo de contaminación ocurre en esta situación.

image.png

**Modelo simple PREGUNTA a)**

```python
QA = 25 # m3/seg
CSA = 9 # mg/l
QW = 10 # m3/seg
CSW = 300 # mg/l

QB = QA + QW
CSB = (QA*CSA + QW*CSW)/QB

print(f"El caudal en el punto B será de {QB:.2f} m3/seg con una concentración de {CSB:.2f} mg/l")
```

**Modelo dinámico**

```python
QA = float(input("Ingrese el caudal del punto A en m3/seg: "))
CSA = float(input("Ingrese la concetración del punto A en mg/l: "))
QW = float(input("Ingrese el caudal del vertimiento en m3/seg: "))
CSW = float(input("Ingrese la concetración del vertimiento en mg/l: "))

QB = QA + QW
CSB = (QA*CSA + QW*CSW)/QB

print(" ")
print(f"El caudal en el punto B será de {QB:.2f} m3/seg con una concentración de {CSB:.2f} mg/l, es decir que desde el punto A aumentará  su carga en {((CSB-CSA)/CSA)*100:.2f} % ")
```

**Modelo básico PREGUNTA b)**

```python
QA = 25 # m3/seg
CSA = 9 # mg/l
RA  = 0.1 #%
QW = 10 # m3/seg
CSW = 300 # mg/l

QB = (QA*RA) + QW
CSB = (QA*CSA*RA + QW*CSW)/QB

print(f"El caudal en el punto B será de {QB:.2f} m3/seg con una concentración de {CSB:.2f} mg/l")
```

# **Ejercicio 3. Eficiencia de tratamiento**

En la empresa que usted trabaja, le han solicitado revisar la eficiencia del proceso de tratamiento de agua residual, que lleva 2 años de funcionamiento, y que se adquirió por su capacidad de remoción de 90% del contaminante en el agua residual. Los datos que ha tomado muestran que al ingresar la concentración del agua residual es de 200 mg/l y la concentración de salida del agua después del proceso es de 30 mg/l

¿Cuál es la eficiencia del proceso de remoción en el tratamiento de agua residual en la empresa?

image.png

```python
C1 = 120 # mg/l
C2 = 55  # mg/l

Eficiencia = ((C1-C2)/C1)*100

print(f"El sistema de tratamiento de agua residual ha demostrado tener una eficiencia de {Eficiencia:.1f}%")
```

# **Ejercicio 4. Eficiencia de tratamiento**

Suponga que debe revisar la eficiencia del proceso de tratamiento de agua residual en la empresa, que tiene una remoción de 80% del contaminante en el agua residual. Si ingresa una carga contaminante con concentración en el agua residual de 200 mg/l
¿Cuál es la concentración del agua después del proceso de remoción en el tratamiento de agua residual en la empresa?

image.png

```python
C1 = 350 # mg/l
Eficiencia = 65 #%

C2 = C1 - (Eficiencia / 100) *C1

print(f"Para un sistema de tratamiento que tiene una eficiencia de {Eficiencia}%, si ingresa una concetración de {C1:.2f} mg/l se obtendra una concentración final del contaminante de {C2:.2f} mg/l")
```

# **QUIZ**

En la ciudad de Villavicencio, están situadas dos molinos de arroz, que tienen vertimientos sobre el cauce del río  Guayuriba, tal y como se observa en la figura 1. Se evidenció que la calidad del agua, antes de los vertimientos, tiene una concentración de 2 mg/l de nitratos y  un caudal de 33 m³/s. Por otro lado, se midió en cada uno de los vertimientos de las industrias A y B, las siguientes concentraciones del contaminante y caudales, respectivamente;  120 mg/l y 338 mg/l con 10 m³/s y 33 m³/s. Asumiendo que ocurre una mezcla completa. Determine:

a) La concentración de nitratos en el agua después de vertimiento del molino B.
b) Indicar ¿Qué ocurrirá si en época de verano se disminuye el caudal del río un 20%?
c) Describir que tipo de contaminación ocurre en esta situación.


```python
# CARGA CONTAMINANTE w = Q * C = (W = Carga contaminante, Q = Caudal, C = Concentración del contaminante)

Qi = 33*0.8 #m3/s
Ci = 2 #mg/litro
QA = 10 #m3/s
CA = 120 #mg/litro
QB = 33 #m3/s
CB = 388 #mg/litro

Qf = Qi + QA + QB
Cf = (Qi*Ci + QA*CA + QB*CB)/Qf

print(f"El caudal despues de los vertimientos es de {Qf:.2f} m3/s con una concentración de {Cf:.2f} mg/l, quiere decir que va aumentar su carga contaminante un {((Cf-Ci)/Ci)*100:.2f}%")
```

```python
# Datos de la empresa láctea
caudal_lactea = 410  # litros/segundo
concentracion_dbo_lactea = 21  # mg/litro

# Datos de la empresa cárnica
caudal_carnica = 0.15  # litros/segundo
concentracion_dbo_carnica = 60000  # mg/litro

# Segundos en un día
segundos_dia = 24 * 60 * 60

# Cálculo de la carga de DBO en kg/día para cada empresa
carga_dbo_lactea = caudal_lactea * concentracion_dbo_lactea * segundos_dia / 1000
carga_dbo_carnica = caudal_carnica * concentracion_dbo_carnica * segundos_dia / 1000

# Determinar qué empresa tiene la mayor carga contaminante
if carga_dbo_lactea > carga_dbo_carnica:
    print("La industria láctea tiene la mayor carga contaminante con", carga_dbo_lactea, "kg/día.")
else:
    print("La industria cárnica tiene la mayor carga contaminante con", carga_dbo_carnica, "kg/día.")
```
